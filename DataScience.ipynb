An in-depth analysis of sales, customer behavior, and product profitability using the Superstore dataset.


```python
import pandas as pd
import pandas as pd

# Use a different encoding to avoid UnicodeDecodeError
data = pd.read_csv('Data.csv', encoding='ISO-8859-1')  # or try 'cp1252'
```

```python
data.head()  # Display the first few rows of the DataFrame
```

```python
'''Analyze a retail store's sales data to identify:

1.Top-performing products and regions

2. Customer purchasing behavior

3. Seasonal trends and actionable business insights'''
```

```python
df=pd.DataFrame(data)
```

```python
max_sales = df['Sales'].max()
```

```python
max_sales_row = df[df['Sales'] == max_sales]
```

```python
max_sales_row
```

```python
df.sort_values(by='Sales', ascending=False).head(10)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df
```

```python
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year
df['Day'] = df['Order Date'].dt.day
df['Weekday'] = df['Order Date'].dt.day_name()
```

```python
import matplotlib.pyplot as plt

monthly_sales = df.groupby(['Year', 'Month'])['Sales'].sum()
monthly_sales.plot(kind='line', figsize=(10, 5))
plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid()
plt.show()
```
